```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
marvel=pd.read_csv('/content/charcters_stats.csv')
```

```python
marvel.head() #returns a specified number of rows, string from the top.The head() method returns the first 5 rows if a number is not specified.
```

```python
marvel.shape #no. of rows and columns
```

```python
marvel['Alignment'].value_counts() #alignment is a categorical column. frequency of the categories
```

```python
good=marvel[marvel['Alignment']== "good"]
```

```python
good.head()
```

To sort the DataFrame based on the values in a single column, you'll use . sort_values() . By default, this will return a new DataFrame sorted in ascending order. It does not modify the original DataFrame.

```python
good.sort_values(by=['Speed'],ascending=False)
```

```python
max_speed_good=good[good['Speed']==100]
```

```python
max_speed_good
```

```python
min_speed_good=good[good['Speed']==1]
```

```python
min_speed_good
```

```python
good.sort_values(by=['Power'],ascending=False)
```

```python
max_power=good[good['Power']== 100]
```

```python
max_power.shape
```

```python
max_power
```

```python
max_power=good.sort_values(by=['Total'],ascending=False)
```

```python
max_power.head()
```

Data visualization

```python
plt.figure(figsize=(8,5))
plt.bar(list(max_power['Name'])[0:5],list(max_power['Total'])[0:5],color="blue")
plt.show
```

```python
bad=marvel[marvel['Alignment']=="bad"]
```

```python
bad
```

```python
bad.sort_values(by=['Intelligence'],ascending=False).head()
```

```python
bad_max_power=bad.sort_values(by=['Total'], ascending=False).head()
```

```python
bad_max_power
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(bad_max_power['Name'])[0:5],list(bad_max_power['Total'])[0:5],color="red")
plt.show
```
